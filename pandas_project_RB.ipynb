```python
#Method _ removing nulls 
#df_isna is an alias for df_isnull, they will return the same output 
#gps.isna().sum().groupby(['Genres'])

print(gps['Genres'].isnull())
```

```python
grouped = gps.groupby('Genres')

# Iterate over each group
for Genres, group_df in grouped:
    print(f"Genres: {Genres}")
    # Counting null values in each column
    null_counts = group_df.isnull().sum()
    # Printing null values for each column
    print(null_counts)
    print("\n")
```
